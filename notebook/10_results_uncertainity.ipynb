```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gpt = pd.read_csv('../qa_predictions/20_total_documents/gpt/original_position_scores_uncertainity_accs.csv')
reorder = pd.read_csv('../qa_predictions/20_total_documents/gpt_reordered/original_position_scores_uncertainity_accs.csv')
tot = pd.read_csv('../qa_predictions/20_total_documents/gpt_tot/original_position_scores_uncertainity_accs.csv')
```

```python
llama = pd.read_csv('../qa_predictions/20_total_documents/llama/original_position_scores_uncertainity_accs.csv')
reorder_llama = pd.read_csv('../qa_predictions/20_total_documents/llama_reordered/original_position_scores_uncertainity_accs.csv')
tot_llama = pd.read_csv('../qa_predictions/20_total_documents/llama_tot/original_position_scores_uncertainity_accs.csv')
cot_llama = pd.read_csv('../qa_predictions/20_total_documents/llama_cot/original_position_scores_uncertainity_accs.csv')
```

```python
llama_70b = pd.read_csv('../qa_predictions/20_total_documents/llama_70b/original_position_scores_uncertainity_accs.csv')
llama_70b_reorder = pd.read_csv('../qa_predictions/20_total_documents/llama_70b_reordered/original_position_scores_uncertainity_accs.csv')
llama_70b_cot = pd.read_csv('../qa_predictions/20_total_documents/llama_70b_cot/original_position_scores_uncertainity_accs.csv')

llama_70b["model"] = "llama_70b"
llama_70b_reorder["model"] = "llama_70b"
llama_70b_cot["model"] = "llama_70b"
llama_70b["method"] = "Regular"
llama_70b_reorder["method"] = "Reorder"
llama_70b_cot["method"] = "Cot"
```

```python
llama["model"] = "llama"
reorder_llama["model"] = "llama"
tot_llama["model"] = "llama"
cot_llama["model"] = "llama"

llama["method"] = "Regular"
reorder_llama["method"] = "Reorder"
tot_llama["method"] = "Tot"
cot_llama["method"] = "Cot"

df = pd.concat([llama, reorder_llama, tot_llama, cot_llama, llama_70b, llama_70b_reorder, llama_70b_cot])
```

```python
gpt["model"] = "GPT"
reorder["model"] = "GPT"
tot["model"] = "GPT"

gpt["method"] = "Regular"
reorder["method"] = "Reorder"
tot["method"] = "Tot"
df = pd.concat([df, gpt, reorder, tot])
```

```python
sns.lineplot(data=df, x="position", y="accuracy", hue="model", style="method", errorbar="sd")
plt.title("Accuracy vs Position")
# Set legend outside the figure
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
